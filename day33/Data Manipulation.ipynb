## Dropping(Deleting) a column

```python
import pandas as pd
```

```python
titanic_df = pd.read_csv("Titanic_data.csv")
```

```python
titanic_df.drop('PassengerId', inplace = True,axis=1)
```

```python
titanic_df
```

```python
titanic_df.drop([0,2,5],axis = 0, inplace= False).head()
```

```python
last_name = [name.split(",")[0] for name in titanic_df["Name"]]
last_name
```

```python
titanic_df['Last_Name'] = last_name
```

```python
titanic_df.head()
```

```python
initial = [titanic_df["Name"][0].split(".")[0].split(",")[-1].strip() for name in titanic_df["Name"]]
```

```python
titanic_df["Initial"] = initial
```

```python
titanic_df.head()
```

```python
titanic_df.drop(["Initial","Last_Name"],axis = 1)
```

```python
titanic_df
```

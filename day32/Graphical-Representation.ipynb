```python
import numpy as np
import pandas as pd
```

## Bar Graph

```python
titanic_df = pd.read_csv("Titanic_data.csv")
```

```python
titanic_df['Survived'].value_counts().plot(
    kind = 'bar',legend = True,
    title = "Count of survived passangers",grid = True,
    fontsize = 18,
    colormap = "tab20c"
)
```

```python
group_by_gender = titanic_df[["Sex","Survived"]].groupby(by="Sex")
```

```python
group_by_gender.value_counts().plot(kind = "bar",legend = True,colormap = "Dark2")
```

```python
group_by_embarked = titanic_df[['Embarked','Age']].groupby(by=['Embarked'])
group_by_embarked.mean().plot(kind = "bar",colormap = "Dark2")
```

## Pie Chart

```python
titanic_df['Survived'].value_counts().plot(
    kind = 'pie',legend = True,
    title = "Count of survived passangers",grid = True,
    fontsize = 18,
    colormap = "tab20c"
)
```

```python
group_by_gender.value_counts().plot(kind = "pie",colormap = "Pastel1")
```

## Line Graph

**Random Module**
--
Provides several types of functions for generating pseudo-random numbers. Pseudo random means not completely random but generated by deterministic algorithm that produces numbers that apperar to be random.

If you know the seed value and the algorithm used, you can predict the sequence of numbers that will be generated.

```python
import random
```

```python
random.random() #random floating value between between 0 and 1 --> Only [0,1)
```

```python
random.uniform(2,3) #random floating value between given 2 values-->[2,3]
```

```python
random.randint(1,100) #generate random integer from 1 to 100. -->[1,100]
```

```python
random.randint(-4,5)
```

```python
random.randrange(-200,100,10) #generates random number in range 1 to 100 with steps 10..ie...generates in multiple of 10
```

```python
list (range(-20,35,3)) #generates random number in range -20 to 35 with the step of 3 and convert to list
```
